## 토큰화

```python
# 토큰화를 위해 토크나이저와 불용어 다운로드
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stop_words = nltk.corpus.stopwords.words('english')
stop_words.extend([])
stop_words[0:10]
```

```python
len(stop_words)
```

```python
# 토큰화, 표제어 추출 작업
# 불용어에 포함되지 않은 것만 저장
from nltk import wordnet
nltk.download('wordnet')
preprocessed_texts=[]

for title in titles.values:
    tokenized_text = tokenizer.tokenize(title)
    lemmatized_text = [lemmatizer.lemmatize(token) for token in tokenized_text]
    stopped_text = [token for token in lemmatized_text if token not in stop_words]
    preprocessed_texts.append(stopped_text)
```

```python
# 토큰 개수 파악
from collections import Counter
tokens=[]
for text in preprocessed_texts:
    tokens.extend(text)

counted_tokens=Counter(tokens)
top_20=counted_tokens.most_common(20)
top_20
```

## 감정분석

```python
# 긍정/부정 단어 다운로드
from nltk.corpus import opinion_lexicon

nltk.download('opinion_lexicon')

positive_words = list(opinion_lexicon.positive())
negative_words = list(opinion_lexicon.negative())
```

```python
len(positive_words), len(negative_words)
```

```python
positive_words[:10]
```

```python
negative_words[:10]
```

```python
from nltk.corpus import gutenberg
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize

nltk.download('punkt')
```

```python
tokens=[]
for i, t in titles.items():
    token=word_tokenize(t)
    tokens.append([i, token])
```

```python
sentiment_id = []

for token in tokens[:100]: # 다 하면 너무 많을 것 같아서 100개만 수행
    i=token[0]
    sentiment=0
    count=0
    
    for t in token[1]:
        if t in positive_words:
            sentiment += 1
            count += 1
        if t in negative_words:
            sentiment -= 1
            count += 1
            
    if count > 0:
        sentiment_id.append([i,count, sentiment/count])
    else:
        sentiment_id.append([i,0,0])

sentiment_id
```

```python
sentiment_id = []

for token in tokens:
    i=token[0]
    sentiment=0
    count=0
    
    for t in token[1]:
        if t in positive_words:
            sentiment += 1
            count += 1
        if t in negative_words:
            sentiment -= 1
            count += 1
            
    if count > 0:
        sentiment_id.append([i,count, sentiment/count])
    else:
        sentiment_id.append([i,0,0])

sentiment_id[:10]
```

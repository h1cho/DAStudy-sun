### 브루어리 DF 탐색

```python
num_breweries_df = breweries['state'].value_counts().reset_index(name="Number_Breweries")
num_breweries_df
```

```python
import seaborn as sns
%matplotlib inline

plt.figure(figsize=(18, 12))
barplot = sns.barplot(data=num_breweries_df, x='state', y="Number_Breweries")

for bar in barplot.patches:
    barplot.annotate(format(bar.get_height(), ".0f"),
                  (bar.get_x() + bar.get_width() / 2,
                  bar.get_height()),
                  ha='center', va='center',
                  size=12, xytext=(0, 8),
                  textcoords="offset points") 
plt.title("Number of Breweries")
plt.show()
```

### 브루어리의 갯수 확인 (States)

```python
fig, axes = plt.subplots(1, 2, sharey=True ,figsize=(10, 8))
fig.suptitle('Breweries Number')

plot1 = sns.barplot(data=num_breweries_df.head(10), x='state', y='Number_Breweries', ax=axes[0])
axes[0].set_title("Top 10 Breweries")
axes[0].set_xlabel("States")

plot2 = sns.barplot(data=num_breweries_df.tail(10), x='state', y="Number_Breweries", ax=axes[1])
axes[1].set_title("Low 10 Breweries")
axes[1].set_xlabel("States")

for bar in plot1.patches:
    plot1.annotate(format(bar.get_height(), ".0f"),
                  (bar.get_x() + bar.get_width() / 2,
                  bar.get_height()),
                  ha='center', va='center',
                  size=10, xytext=(0, 8),
                  textcoords="offset points") 

for bar in plot2.patches:
    plot2.annotate(format(bar.get_height(), ".0f"),
                  (bar.get_x() + bar.get_width() / 2,
                  bar.get_height()),
                  ha='center', va='center',
                  size=10, xytext=(0, 8),
                  textcoords="offset points") 
plt.show()
```

### 미국에서는 어떤 스타일의 맥주가 인기가 많을까?

* American Style의 비중이 높은 것을 확인할 수 있었습니다

```python
beers.isnull().sum()
```

```python
beers_style_ratio = beers['style'].value_counts(normalize=True).reset_index(name="Percent").head(10)
beers_style_ratio['Percent'] = round((beers_style_ratio['Percent'] * 100), 2)
beers_style_ratio
```

```python
beers_style_ratio['Percent'].sum()
```

#### American Style Beer에 대해서 좀 더 알아보자

```python
beers_american = beers[beers['style'].str.contains('American', na=False)]
beers_american
```

```python
beers_american.isnull().sum()
```

```python
beers_american.describe()
```

```python
beers_american['style'].value_counts(normalize=True).head(10).plot(kind="pie", autopct="%1.1f%%")
plt.title("Top 10 American style Beer percents")
plt.show()
```
